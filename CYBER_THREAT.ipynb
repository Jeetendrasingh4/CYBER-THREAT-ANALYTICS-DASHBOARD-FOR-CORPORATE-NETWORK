### DATA CLEANING

##### Column Names Correction

```python
df= df.rename(columns={
    "Date_Repoted" : "Date_Reported",
    "Threat_Typ" : "Threat_Type",
    "Severtiy" : "Severity",
    "Departmnt" : "Department",
})
```

```python
df.columns
```

##### Checking Null Values

```python
print(df.isnull().sum())
```

##### Handling Null Values

```python
df["Threat_Type"] = df["Threat_Type"].fillna("Unknown")
df["Severity"] = df["Severity"].fillna("Unknown")
df["Department"] = df["Department"].fillna("Unknown")
df["Source_Country"] = df["Source_Country"].fillna("Unknown")
```

##### Checking Duplicate Values

```python
df["Incident_ID"].duplicated().sum()
```

##### Handling Duplicate Values

```python
df.drop_duplicates()
```

### KPI REQUIREMENTS

```python
#Total incident
Total_incident = df["Incident_ID"].count()

#Total Malware attack
Total_malware = (df["Threat_Type"]=="Malware").sum()

#Total Phishing attack
Total_phishing = (df["Threat_Type"]=="Phishing").sum()

#Total Intrusion attemps
Total_intrusion = (df["Threat_Type"]=="Intrusion").sum()

#Total Ransomware attack
Total_ransomware = (df["Threat_Type"]=="Ransomware").sum()

#Total failed login
Total_failed_login = df["Failed_Logins"].sum()

#High severity incident
high_severity_incident = (df["Severity"]=="High").sum()

#average of MTTD(mean time to detect)
avg_MTTD = (df["MTTD_minutes"].sum())/df["MTTD_minutes"].count()

#average of MTTR(mean time to respond)
avg_MTTR = (df["MTTR_hours"].sum())/df["MTTR_hours"].count()

#Total incident Resolved
Total_incident_resolve = (df["Status"]=="Resolved").sum()

# Display

print("Total Incident :",Total_incident)
print("Total Malware Attacks :",Total_malware)
print("Total Phishing Attacks :",Total_phishing)
print("Total Intrusion Attacks :",Total_intrusion)
print("Total Ransomware Attacks :",Total_ransomware)
print("Total failed logins :",Total_failed_login)
print("Total High severity incidents :",high_severity_incident)
print("Average of MTTD(mean to time detect) :",avg_MTTD)
print("Average of MTTR(mean to time respond) :",avg_MTTR)
print("Total Incident resolved :",Total_incident_resolve)
```
